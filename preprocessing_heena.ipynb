```python
import konlpy
```

```python
# Okt 토큰화 & 태깅
from konlpy.tag import Okt

def tokenize(text):
    Tag = []
    okt = Okt()
    tokens = okt.morphs(text)
    for token in tokens:
        Tag += okt.pos(token)
    return Tag #태깅된 토큰들 
```

```python
df['pos'] = df['answer'].apply(lambda x: tokenize(x))
```
